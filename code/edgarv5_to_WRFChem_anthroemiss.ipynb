This notebook prepares the edgarv5 emissions to be used in WRF-Chem preprocessing tool anthro-emiss by adding date and datetime variables and units attribute needed.

```python
import xarray as xr
import os
import numpy as np
import pandas as pd
```

```python
#define input and output paths
ed_pth = '/geos/d21/s1878599/edgarv5_process/monthly_all_sectors_with_tot/'
save_dir = '/geos/d21/s1878599/edgarv5_process/edgar2015_mozart_wrfchem/'

#create save dir if doesn't exist.
if not os.path.isdir(save_dir):
    !mkdir -p $save_dir
```

```python
ds=xr.open_dataset(ed_pth +'monthly_v50_2015_CO_.0.1x0.1.nc')
```

## Add monthly date and datesec variables and units attribute

```python
for f in os.listdir(ed_pth):
        sp=f.split('_')[3]
        print(sp)
        s=xr.open_dataset(ed_pth+f)
        
        #clean variables attributes and add units attributes
        for sec in list(s.keys()):
            s[sec].attrs['units']='kg m-2 s-1'
        #add datetime variables.
        s=s.assign({'date': xr.DataArray(data=np.array([20150101, 20150201, 20150301, 20150401, 20150501, 20150601,
           20150701, 20150801, 20150901, 20151001, 20151101, 20151201]),dims=["time"])})
        s=s.assign({'datesec': xr.DataArray(data=np.zeros(12),dims=["time"])})
        #save
        s.to_netcdf(save_dir +'edgar_v5_2015_'+ sp +'_0.1x0.1.nc',format='NETCDF3_64BIT')
```
